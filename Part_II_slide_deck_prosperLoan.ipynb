> Un pret est le fait pour un établissement de crédit, de proposer une solution de financement, en mettant à disposition des fonds à un bénéficiaire, sans en exiger le remboursement immédiat. Dans ce document, nous allons étudier un ensemble de données de pret, fournit par Prosper. 

> Ce jeu de données comprend un ensemble 113 937 prêts avec 81 variables sur chaque prêt, y compris le montant du prêt, le taux de l'emprunteur (ou taux d'intérêt), le statut actuel du prêt, le revenu de l'emprunteur, et bien d'autres. 

> **Exploitation**: Dans le but de retirer des informations utiles de ce jeu de données, nous nous posons les questions suivantes :᷆

> <ul>
    <li>Quels facteurs affectent le statut du résultat d'un pret ? </li>
    <li>QU'est ce qui affecte l'APR ou le taux d̀intéret de l'emprunteur ?</li>
    <li>Existe-t-il des différences entre les prets en fonction de l'importance du montant initialement emprunté ?</li>
</ul>


Afin de repondre a ces questions, nous avons exploré notre dataset à travers plusieurs caractéristiques. 
Les caracteristiques d'interet principales étant :
> <ul>
    <li>LoanStatus -- statut du pret</li>
    <li>BorrowerAPR -- taux annuel effectif global de l'emprunteur</li>
    <li>BorrowerRate -- Taux d'interet de l'emprunt pour ce pret</li>
    <li>LenderYield -- Rendement du preteur </li>
    <li>LoanOriginalAmount -- montant initial du pret</li>
  </ul>

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df_loan_clean = pd.read_csv('df_loan_clean.csv')
````

## Visualisation 1

> En étudiant les paramètres de notre jeu de données en fonction de leur fréquences, nous avons déterminé que la tranche de revenus est une caractéristique d'intéret pour notre étude.
> Nous essayons de déterminer quels sont les tranches de revenus qui demandent le plus de pret.

````python
sorted_counts = df_loan_clean.IncomeRange.value_counts()
colors = ['midnightblue','mediumblue','royalblue','cornflowerblue','slateblue', 'paleturquoise','lavender','azure']
plt.figure(figsize =(4,19))
plt.pie(sorted_counts,  startangle = 90, counterclock = False, colors =colors)
plt.title('Tranche de revenus', fontweight = 'bold', fontsize = 'xx-large', y= 1.04)
plt.legend(sorted_counts.index, bbox_to_anchor=(1.3, 0.7), ncol=2)
plt.show();
````

> On constate que les personnes à revenus moyen donc compris entre __25 000 et 75 000 dollars__ sont les plus demandeuses de pret.

## Visualisation 2

> Partant du constat précédemment effectué, nous observons la fréquence des status de pret en fonction des différentes tranches de revenus.
> Nous essayons de déterminer quels sont les tranches de revenus dont les prets sont en cours de rembousement ou ont déjà été remboursés.

````python
plt.figure(figsize =[12,5])
plt.title('Statut de pret par tranche de revenu', fontweight = 'bold', fontsize = 'xx-large', y= 1.04)
plt.xlabel('LoanStatus', fontweight = 'semibold', fontsize = 'large', labelpad = 20)
plt.ylabel('count', fontweight = 'semibold', fontsize = 'large', labelpad = 20)
plt.legend('IncomeRange')
#sb.color_palette('rocket', as_cmap =True)
sb.countplot(data = df_loan_clean, x ='LoanStatus',  hue = 'IncomeRange', palette ='Blues_r');
````

> On constate également que ces tranches de revenus sont de loin les plus présente dans les catégories de pret complétés ou en cours de remboursement.

## Visualisation 3

> Nous avons fait une observation du statut d̀un pret en fonction du taux annuel attribué à l'emprunteur ainsi que sa profession.
> Ce visuel a pour but de déterminer si la ou les professions qui pouraient avoir une revalorisation de leur taux. 

````python
LoanRentable = df_loan_clean[df_loan_clean.LoanStatus.isin(['Completed','Current','FinalPaymentInProgress'])]
Loan10occ = LoanRentable[LoanRentable.Occupation.isin(['Scientist', 'Architect', 'Engineer - Electrical',
                  'Attorney', 'Pilot - Private/Commercial',
                  'Computer Programmer', 'Engineer - Chemical', 'Pharmacist',
                  'Doctor', 'Judge'])]
occ10_type = pd.CategoricalDtype(categories = ['Scientist', 'Architect', 'Engineer - Electrical',
                 'Attorney', 'Pilot - Private/Commercial',
                  'Computer Programmer', 'Engineer - Chemical', 'Pharmacist',
                  'Doctor', 'Judge'], ordered =True)
Loan10occ.Occupation = Loan10occ['Occupation'].astype(occ10_type)
plt.figure(figsize =(10, 8))
plt.title('Statut de pret par occupation et APR', fontweight = 'bold', fontsize = 'xx-large', y= 1.04)
plt.xlabel('BorrowerAPR', fontweight = 'semibold', fontsize = 'large', labelpad = 20)
plt.ylabel('Occupation', fontweight = 'semibold', fontsize = 'large', labelpad = 20)
sb.barplot(data = Loan10occ , y = 'Occupation', x = 'BorrowerAPR',orient ='h', hue = 'LoanStatus',  palette ='rocket');
````

> De nos observations on constate que le taux annuel de l'emprunt est plus élevé pour les __ingénieurs électriciens__ en phase final du processus de remboursement. Donc améliorer leur taux pourrait les aider  à rembourser plus vite.
